# Clean Data

```python
import pandas as pd
```

```python
df = pd.read_csv('Retail_Sales_and_Retail_Use_Business_Registrations.csv')
```

```python
missing_summary = df.isnull().sum().sort_values(ascending=False)
```

```python
missing_summary
```

```python
df.drop(columns=['Date Permit was Cancelled','Consolidated Number','Department City Code','Department City Code','Description of type of NAICS code','NAICS code','Business Zip Code','Address of the Business','Business Location County Number','Retail Permit Number','Registration County Number'],inplace=True)
```

```python
df  
```

```python
city_to_county = (
    df[df['County Name'] != 'Unknown'][['City Location of Business', 'County Name']]
    .dropna()
    .drop_duplicates()
    .set_index('City Location of Business')['County Name']
    .to_dict()
)
```

```python
df['County Name'] = df.apply(
    lambda row: city_to_county.get(row['City Location of Business'], row['County Name']),
    axis=1
)
```

```python
print("Remaining 'Unknown' County Names:", (df['County Name'] == 'Unknown').sum())
```

```python
df.loc[:, 'County Name'] = df['County Name'].fillna('Unknown')
df.loc[:, 'State'] = df['State'].fillna('IA')
df = df[df['City Location of Business'].notnull()]
```

```python
df.to_csv("Cleaned_Business_Registrations.csv", index=False)
```
